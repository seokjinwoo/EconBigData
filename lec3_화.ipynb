```python
# scatter plot. 산포도
# 두 변수의 관계를 볼 때 산포도를 가장 많이 사용함

from numpy.random import standard_normal
import random
import matplotlib.pyplot as plt

# random.seed(42)

x = standard_normal(100)
y = standard_normal(100)

# x, y의 상관관계
plt.figure(figsize=(4,3))
plt.scatter(x,y, marker='o')
plt.show()
```

```python
# 여러 개의 그림을 한 그림 안에
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc=0, scale=2, size=100)
u = np.random.normal(loc=0, scale=1, size=100)

y = 1 + 0.5*x + u

# 2개의 그래프:
# x의 히스토그램, x & y의 상관관계

fig, ax = plt.subplots(2,1, figsize=(3,6))

ax[0].hist(x)
ax[0].set_title('x의 히스토그램')

ax[1].scatter(x,y, marker='o')
ax[1].set_title('x와 y의 상관관계')

plt.show()

# 각 그래프의 타이틀을 넣어보시오
# x의 히스토그램, x와 y의 상관관계


# 그림 배치를 세로로

```
