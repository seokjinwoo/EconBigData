# 단순회귀분석(simple regression analysis)

- $y$: 종속변수, 피설명변수
- $x$: 독립변수, 설명변수.

- 설명변수가 1개인 모형: 단순회귀모형

$$
y_i = \beta_0 + \beta_1 x_i + e_i , i=1, \dots, n
$$

- 오차 제곱의 합

$$
\sum_{i=1}^n e_i^2 = \sum_{i=1}^n  (y_i - (\beta_0 + \beta_1 x_i) )^2
$$

- 목적함수(objective function), 손실함수(loss function)

- **OLS(ordinary least square)**: 최소자승추정법

```python
import numpy as np
import pandas as pd

np.random.seed(123456)

# y = a + b x + e
e = np.random.normal(0,100,1000)  # error term
x = np.random.normal(1,3,1000)  # 설명변수

y = 1 + 1*x + e                 # 피설명변수

df = pd.DataFrame({'x':x, 'y':y})
# df.head()

# x와 y의 산포도와 회귀선을 그려보시오.

plt.figure(figsize=(5,3))
sns.regplot(x='x', y = 'y', data=df, ci=False)
plt.show()
```

```python
import statsmodels.formula.api as smf

# define the regression model
model = smf.ols('y ~ x', data=df).fit()
print(model.summary())
```
