## 기대값(expectation)

- 평균, 기대값, 기대치, 평균치, 평균값
- mean, expectation

- 분포의 중심, center, location
- typical unit

### 이산확률변수의 기대값

- 이산확률변수: countable 값을 갖는 확률변수

- 평균(mean)
$$
E[X] = \sum_{i=1}^n x_i \Pr(X = x_i) = \mu
$$

- 분산(variance): dispersion, 얼마나 퍼져 있는지
$$
var(X) = E[(X-E(X))^2] = \sum_{i=1}^n (x_i - E(X))^2 \Pr(X = x_i) = \sigma^2
$$


### 연속확률변수의 기대값

- 불가산(uncountable) 의 값을 갖기 때문에 pmf를 이용하는 대신 pdf 를 이용해서 정의

- 평균
$$
E[X] = \int_{A} x f(x) df = \mu
$$

  - $A$ 는 확률변수 $X$가 정의되는 구간

- 분산
$$
var(X) = \int_{A} (x-\mu)^2 f(x) df = \sigma^2
$$


### 중앙값(median)

- 중앙값도 중심을 나타내는 또 하나의 통계량

- 이 값의 아래로 50% 있고, 이 값의 위로 50%

- 분포함수 $F(\cdot)$에 대해서
$$
F(med) = 0.5
$$













```python
# chi2 분포

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# parameters
n = 6 # 자유도
x = np.linspace(0, 20, 200)
y = chi2.pdf(x, n)

# mean
mean = chi2.mean(n)

# median
median = chi2.ppf(0.5,n)
# print(median)

# add a median line to the figure with blue vertical line
plt.figure(figsize = (5,3))
plt.fill_between(x, y, color = 'gray', alpha = 0.3)
plt.axvline(x = mean, color = 'red', linestyle = '--', label = 'mean')
plt.axvline(x = median, color='blue', linestyle = '-', label = 'median')

plt.legend()
plt.show()
```

## Box-plot


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
y1 = np.random.normal(1, 2, 1000)
y2 = np.random.normal(3, 1, 1000)
y3 = np.exp(np.random.normal(0,1,1000))   # 로그-정규분포 LN(0,1) 의 box-plot 을 더해보시오

plt.figure(figsize=(5,3))
plt.boxplot([y1,y2,y3], tick_labels = ['$N(1,2^2)$','N(3,1)','LN(0,1)'])
plt.show()
```

# 다변량(multivariate) 분석

- 확률변수 2개이상의 상호 관계에 관심이 있음

- 수요함수: P(가격) → Q(수요량)
- 물가: M(통화) → prices(물가)
- 재정 G(정부지출) → C(소비)

- 확률변수 $X$와 $Y$의 관계
  - **인과관계**(causal relation)일 수도 있고, 단순 상관관계(corrleation)일 수도 있음

  - Economics science

- 공분산(covaraince)

$$
cov(X,Y) = E[ (X - \mu_X)  (Y - \mu_Y)  ]
$$


- 공분산을 표준화 시켜서 -1 과 1 사이의 숫자로 만듦

- 상관계수(corrleation coefficient)

$$
\rho_{X,Y} = \frac{ cov(X,Y) }{ \sqrt{var(X)}  \sqrt{var(Y)}}
$$









```python
import pandas as pd
import seaborn as sns # seaborn
import matplotlib.pyplot as plt

df = pd.read_csv('/content/mid_final_scores.txt', sep = '\t')
# df.head()

# 가로축에 mid(중간고사 성적), 세로축(기말고사 성적)
# 산포도 그려보시오.

plt.figure(figsize = (3,2))

sns.scatterplot(data = df, x = 'mid', y = 'final')
sns.regplot(data=df, x='mid', y='final', scatter=False, ci=False)

plt.show()
```

```python
from scipy.stats import pearsonr

corr_coef, _ = pearsonr(df['mid'], df['final'])
# print(corr_coef)
print(f"중간 기말 상관계수는 {corr_coef: .2f}")



```
