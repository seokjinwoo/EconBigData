## 확률변수와 분포

- 확률변수(random variable) $X$ (Latex이라고 하는 과학에서 많이 사용되는 타이셋 언어)
- 상황에 따라서 확률변수란 이 값도 가질 수 있고, 저 값도 가질 수 있는 그런 변수

- 확률분포(probability distribution): 확률변수의 특성 중 하나
$$
f(x) = \Pr(X=x)
$$


- 우리의 관심은 모집단 분포(population distribution)에 있음

- 표집(sampling)을 통해 표분 분포(sample distribution)을 분석

- 표본이 모집단을 잘 대표하려면
  - 임의 표본(random sample) 이어야 함
  - 충분히 많은 표본이 확보되어야 함

## 히스토그램

- 확률분포를 묘사하는 다양한 방법이 존재함

- 표로 만들거나
- 시각화해서 **히스토그램**(histogram)
  - 히스토그램의 세로축은 빈도(frequency)가 아니라 밀도(density)
  - 밀도란 히스토그램 아래 면적이 늘 1이 되도록 조정한 것
  - 빈도는 유용하기는하지만 정확하지는 않음




```python
x = [
    2.4, 3.9, 7.1, 7.3, 7.8, 7.9, 7.9, 8.9, 9, 9.5,
    9.6, 9.8, 10, 10.3, 10.5, 12.4, 12.5, 13, 13.2, 13.6
]

print(x)
```

## 특정 확률분포로부터의 난수 추출

- 이 분포로부터 임의표집을 해서 표본이 구성

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# seed 고정
np.random.seed(123456)


# pdf of standard normal distribution
xmin = -4
xmax = 4
x_vals = np.linspace(xmin, xmax, 100)
y_vals = stats.norm.pdf(x_vals)

# n = 10, 100, 1000, 10000
# 4장의 그림을 2X2 의 그래프 한 장으로 그리시오.

n = 1000
x = np.random.normal(size=n)
# print(x)


#
plt.figure(figsize=(3,2))

plt.hist(x, bins='auto', density = True, rwidth=0.5, label='hist')
plt.plot(x_vals, y_vals, label='pdf')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# seed 고정
np.random.seed(123456)

# pdf of standard normal distribution
xmin = -4
xmax = 4
x_vals = np.linspace(xmin, xmax, 100)
y_vals = stats.norm.pdf(x_vals)

# n = 10, 100, 1000, 10000
ns = [10, 100, 1000, 10000]

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(6, 4))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, n in enumerate(ns):
    x = np.random.normal(size=n)

    # Plot histogram and PDF in the current subplot
    axes[i].hist(x, bins='auto', density=True, rwidth=0.5, label='hist')
    axes[i].plot(x_vals, y_vals, label='pdf')
    axes[i].set_title(f'n = {n}')
    axes[i].legend()

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

---

- $n$ 이 커질수록 표본의 분포가 모집단의 분포로 수렴하게 됨(대수의 법칙)

---

- 표준정규분포를 따르는 $X$의 제곱의 분포 $X^2$
- $X^2$의 히스토그램

```python
# 표준정규분포에서 1000개를 뽑고
# 제곱해서
# 히스토그램 그리시오

x = np.random.normal(size=1000)
y = x**2

plt.figure(figsize=(3,2))
plt.hist(y, bins='auto', density = True, rwidth=0.5)
plt.show()
```

```python
chi = np.random.chisquare(df=100, size=1000)
plt.figure(figsize=(3,2))
plt.hist(chi, bins='auto', density = True, rwidth=0.5)
plt.show()
```
