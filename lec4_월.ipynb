## 확률변수와 분포

- 확률변수 $X$ (Latex: 과학에서 많이 사용하는 타이셋 언어)
- 어떤 값도 가질 수 있는 (확률)변수
- 확률분포 probability distribution
$$
f(x) = \Pr(X=x)
$$

- **모집단**(population): population distribution
- **표본**(sample): 대표성 sample distribution

- 대표성을 구성하는 가장 중요한 요소는 임의성(random sample)
- 충분히 많은 표본수가 필요함







---

### 히스토그램

- 분포를 어떻게 나타낼 것인가?

- 분포를 표시하는 방법은 다양함
  - 제일 쉬운 것이 **표**를 작성
  - 가장 많이 쓰이는 방법은 **히스토그램**(histogram)

- 히스토그램은 $X$를 같은 크기의 구간으로 나누고, 구간안에 들어가는 관측치의 갯수를 센다.
  - 히스토그램의 세로축은 밀도(density)로 하는 것이 정답
  - 빈도(frequency)는 유용하나 정확하지는 않음




```python
x =[
    2.4, 3.9, 7.1, 7.3, 7.8, 7.9, 7.9, 8.9, 9, 9.5,
    9.6, 9.8, 10, 10.3, 10.5, 12.4, 12.5, 13, 13.2, 13.6
]

# print(x)
```

```python
# !pip install koreanize-matplotlib
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure(figsize=(3, 2))
plt.hist(x, bins='auto', density=True, rwidth=0.9)
plt.show()
```

### 특정 확률분포로부터의 난수 추출

- 잘 알려진 확률분포(대표적으로 정규분포 normal distribution)
- 그 분포로부터 난수를 추출한다(random sapmple)



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# set seed
np.random.seed(123456)


n_sim = 10000 # 100, 1000, 10000, 100000 에 대해서 각각 그리고
              # 2X2 그래프에 넣으시오.

x = np.random.normal(size=n_sim)

xmin = -4
xmax =  4
x_vals = np.linspace(xmin, xmax, 100)
y_vals = stats.norm.pdf(x_vals)

# x-축: N(0,1)인 난수, y축: 밀도, 타이틀: 히스토그램
plt.figure(figsize=(6, 3))

plt.hist(x, bins='auto', density=True, rwidth=0.5, label='histogram')
plt.plot(x_vals, y_vals, 'r', label='pdf ')

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import koreanize_matplotlib

# set seed
np.random.seed(123456)

n_sim_list = [100, 1000, 10000, 100000]

xmin = -4
xmax =  4
x_vals = np.linspace(xmin, xmax, 100)
y_vals = stats.norm.pdf(x_vals)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes = axes.flatten()

for i, n_sim in enumerate(n_sim_list):
    x = np.random.normal(size=n_sim)
    axes[i].hist(x, bins='auto', density=True, rwidth=0.5, label='histogram')
    axes[i].plot(x_vals, y_vals, 'r', label='pdf ')
    axes[i].set_title(f'n_sim = {n_sim}')
    # axes[i].legend()

plt.tight_layout()
plt.show()
```

---

- sample distribution(histogram) 은 population distribution으로 수렴하게 됨.

- 표준정규분포에서 n_sim = 1000 인 $X$를 추출하고
