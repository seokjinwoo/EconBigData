```python
from numpy.random import standard_normal
import random
import matplotlib.pyplot as plt

# set seed for random number generation
random.seed(1234)

# x,y 난수 생성
x = standard_normal(100)
y = standard_normal(100)

# x,y와 상관관계
plt.figure(figsize=(4,3))
plt.scatter(x,y, marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
# 여러 개의 그림을 넣어야 될 때가 있음

import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc=0, scale=2, size=100)
u = np.random.normal(loc=0, scale=1, size=100)

y = 1 + 0.5*x + u

fig, ax = plt.subplots(1,2, figsize=(6,3))

ax[0].hist(x)
ax[0].set_title('x의 히스토그램')

ax[1].scatter(x,y)
ax[1].set_title('x와 y의 산포도')

plt.tight_layout()

plt.show()

# 왼쪽 그림에 "x의 히스토그램", 오른쪽 그림에 "x와 y의 산포도" 제목을 다시오
```
