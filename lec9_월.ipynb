# 단순회귀분석(simple regression analysis)

- 설명변수 $x$ 가 1개인 경우

$$
y_i = \beta_0 + \beta_1 x_i + e_i, i=1, \dots, n
$$

- 오차의 제곱의 합

$$
\sum_{i=1}^n e_i^2 = \sum_{i=1}^n \left(y_i -(\beta_0 + \beta_1 x_i) \right)^2
$$

- 목적함수(objective function), 손실함수(loss function)

- **OLS(Ordinary Least Square)** 최소자승추정법



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/galton.csv')


plt.figure(figsize=(5,3))

sns.regplot(x='father', y='height', data=df, ci=False)

plt.show()
```

```python
import numpy as np
import pandas as pd

# seed
np.random.seed(123456)

# y = a + b x + e
e = np.random.normal(0,2,1000)  # error term
x = np.random.normal(1,3,1000)  # explanatory variable

y = 1 + 1*x + e                 # explained variable

df = pd.DataFrame({'x':x, 'y':y})
df.head()

# X와 y 간의 scattergram 을 그려보시오
import seaborn as sns

plt.figure(figsize=(5,3))
sns.regplot(x='x', y='y', data=df, ci=False)
plt.show()
```

```python
# OLS: statsmodels

import statsmodels.formula.api as smf

# define the regression model
model = smf.ols(formula='y ~ x', data = df).fit()
print(model.summary())
```

```python
# auto.csv 를 읽어 들이고

df = pd.read_csv('/content/auto.csv')
# df.head()

# x: price, y: mpg 의 산포도
# 회귀라인을 그려보시오
sns.regplot(x='price', y='mpg', data=df, ci=False)
plt.show()

# mpg = a + b price + e 모형 추정해보시오.
model = smf.ols(formula='mpg ~ price', data = df).fit()
print(model.summary())



```
